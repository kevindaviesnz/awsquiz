````python
--- http://serverless.com/framework/docs/getting-started

$ npm install -g serverless

---  Create a new serverless project
$ serverless 
-- or
$ mkdir <dir>
$ cd <dir>
$ serverless create -template aws-python3

-- Move into the newly created directory
cd your-service-name

-- Deploy project
$ sls deploy
-- or
-- https://www.serverless.com/framework/docs/providers/aws/cli-reference/invoke
$ serverless invoke local --function hello

-- https://us-east-1.console.aws.amazon.com/lambda/home?region=us-east-1#/functions/serverlesstest-dev-hello?tab=code
-- arn:aws:lambda:us-east-1:881617337234:function:serverlesstest-dev-hello

-- Developing on the cloud
$ sls deploy function -f hello
$ serverless invoke -f hello --log

$ sls remove

````
